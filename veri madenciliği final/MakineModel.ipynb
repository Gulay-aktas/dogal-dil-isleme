
veri eğitimi için train_predict foksiyonunu olusturduk. Farklı algoritmalar denediğimizde eğitmek için sürekli aynı kodları yazmak zorunda kalamayacağız.

Normalleştirme dediğimiz yöntem değişkenlerin 0 ve 1 arasındaki değerlere atanmasıdır.Mesela bir değer 12 iken aynı kolonda 2458 gibi bir değer var. Aradaki fark yüzünden eğitim sağlıklı olamayabili.r Datamızdatki verileri normalizasyonma sokarak bu değerleri 0 ile 1 arasına indirgeriz. Bu yöntem sayesinde rakamların kendileri arasındaki otorite  ve büyük farklar kırılır.

 

Fonksiyomuz dataset, hata matrisi, kesinlik vb ve classifier verilerinden oluşmakta.

```python
def train_predict(dataset, classifier): 
   
    #Verimizin %75 ini eğitim %25 ini test içi ayırdık
    X_train, X_test, y_train, y_test = train_test_split(X, y,
        test_size = 0.25,random_state = 5) 
      #random stateye verdğimiz değer sürekli aynı yerden verileri böler 
        #Bundan dolayı program her çalıstırılıdğında sonuçlar aynı cıkar.

    sc = Normalizer()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
   
    classifier.fit(X_train, y_train,)

    y_pred = classifier.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    pr = precision_score(y_test, y_pred,zero_division=0) 
    rc = recall_score(y_test, y_pred,zero_division=0)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred,zero_division=0)

    print("confusion matrix",cm)
    print("precision",pr)
    print("recall",rc)
    print("accuracy", accuracy)
    print("f1 score",f1)

 
    return  classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion = 'entropy',random_state=1)
classifier = train_predict(dataset,clf) 
#datasetimizi eğitmek içintrain_predict foksiyonuna yolluyoruz.
# DecisionTreeClassifier ile eğitilecek
#Aşağıda dt ile eğitilen modele ait confusion matrix , precision(kesinlik), recall(duyarlılık) ,accuracy(doğruluk) ve  f1_scorları mevcut.
```
